#### 사진에서 음영 영역(gray scale) 추출하기

```julia
Xgrayvalues = Float64.(Xgray)
```

```julia
SVD_V = svd(Xgrayvalues)
```

```julia
norm(SVD_V.U*diagm(SVD_V.S)*SVD_V.V' - Xgrayvalues)
```

```julia
# use the top 4 singular vectors/values to form a new image
u1 = SVD_V.U[:,1]
v1 = SVD_V.V[:,1]
img1 = SVD_V.S[1]*u1*v1'

i = 2
u1 = SVD_V.U[:,i]
v1 = SVD_V.V[:,i]
img1 += SVD_V.S[i]*u1*v1'

i = 3
u1 = SVD_V.U[:,i]
v1 = SVD_V.V[:,i]
img1 += SVD_V.S[i]*u1*v1'

i = 4
u1 = SVD_V.U[:,i]
v1 = SVD_V.V[:,i]
img1 += SVD_V.S[i]*u1*v1'
```

```julia
Gray.(img1)
```

- 여전히 원본과는 거리가 있는 사진이므로 
- SVD 요소 1~100열까지를 이용 복원

```julia
i = 1:100
u1 = SVD_V.U[:, i]
v1 = SVD_V.V[:, i]
img1 = u1 * spdiagm(0 => SVD_V.S[i])*v1'
```

```julia
Gray.(img1)
```

- 원본과 비슷하게 보여도 원본과는 차이가 있음
- 이는 norm 계산값의 차이를 통해 알 수 있음

```julia
norm(Xgrayvalues-img1)
```
